```python
import json
import pickle
from text_processing_utils import *
```

```python
selected_books = set()
file = "meta_Books.jsonl"
with open("data/amazon_data/" + file, 'r') as fp:
    c = 0
    for line in fp:
        d = json.loads(line.strip())
        l = d['categories']
        for e in l:
            if "Literature" in e or "Fiction" in e: # filter by book category
                selected_books.add(d['parent_asin'])
                break
print("Total # of books:", len(selected_books))
```

```python
file = "Books.jsonl"
ratings = []
titles = []
reviews = []
books = []
timestamps = []
upvotes = []
with open("data/amazon_data/" + file, 'r') as fp:
    for line in fp:
        d = json.loads(line.strip())
        if d['parent_asin'] in selected_books:
            books.append(d['parent_asin'])
            ratings.append(d['rating'])
            titles.append(d['title'])
            reviews.append(remove_punctuation_regex(d['title'] + " " + d['text']))
            timestamps.append(d['timestamp'])
            upvotes.append(d['helpful_vote'])
print("Total # of documents (reviews):", len(ratings))
```

```python
pickle.dump(ratings, open("data/amazon_data/ratings.pickle", "wb"))
pickle.dump(titles, open("data/amazon_data/titles.pickle", "wb"))
pickle.dump(reviews, open("data/amazon_data/reviews.pickle", "wb")) # upload it to my Google Drive
pickle.dump(books, open("data/amazon_data/books.pickle", "wb"))
pickle.dump(timestamps, open("data/amazon_data/timestamps.pickle", "wb"))
pickle.dump(upvotes, open("data/amazon_data/upvotes.pickle", "wb"))
```

```python
all_vocab = set()
for r in reviews:
    wl = tokenize(r)
    all_vocab.update(wl)
all_vocab_dict = {}
c = 0
for v in all_vocab:
    all_vocab_dict[v] = c
    c += 1
pickle.dump(all_vocab_dict, open("data/amazon_data/all_vocabulary_dict.pickle", "wb")) # upload it to my Google Drive
```
