When merging the large number of word counts by document files, Google Colab had a weird error of becoming frozen after 40-ish files have been uploaded. So we need a special way of dealing with this issue: we reset the Jupyter Notebook environment once every 30 files have been loaded.

```python
all_wc_files = sorted(glob.glob("/content/drive/MyDrive/fragments/wc_new*.pickle"))[::-1]
lll = 30 # block size
ll = len(all_wc_files) // lll + 1
if len(all_wc_files) % lll == 0:
    ll -= 1
```

```python
l = 0
while l < ll:
    from google.colab import drive
    drive.mount('/content/drive', force_remount=True)
    import pickle
    import glob
    import numpy as np
    all_wc_files = sorted(glob.glob("/content/drive/MyDrive/fragments/wc_new*.pickle"))[::-1]
    i = lll*l
    end_indx = i + lll
    if l == ll-1:
        end_indx = len(all_wc_files)
    wc_files = ["/content/drive/MyDrive/fragments/wc_new{}.pickle".format(j) for j in range(i, end_indx)]
    reviews_bow = []
    for fn in wc_files:
        print(fn.split("/")[-1].split(".")[0][6:])
        wc = pickle.load(open(fn, "rb"))
        reviews_bow.extend(wc)
    pickle.dump(reviews_bow, open("/content/drive/MyDrive/fragments/wc_merged{0}_{1}.pickle".format(i,end_indx-1), "wb"))
    l += 1
    # only keep l, ll and lll after each iteration, but clean up everything else
    %reset_selective -f ^(?!l).*$
    %reset_selective -f .*(?<!l)$
    %reset_selective -f ^l(?!l)l$
    %reset_selective -f ....
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
import pickle
import glob
import numpy as np
all_wc_files = glob.glob("/content/drive/MyDrive/fragments/wc_new*.pickle")
reviews_bow = []
for j in range(ll):
    start_indx = j * lll
    end_indx = (j+1) * lll
    print(start_indx, end_indx-1)
    if j == ll - 1:
        end_indx = len(all_wc_files)
    wc = pickle.load(open("/content/drive/MyDrive/fragments/wc_merged{0}_{1}.pickle".format(start_indx, end_indx-1), "rb"))
    reviews_bow.extend(wc)
```

```python
N = len(reviews_bow)
n = 0
c = 0
vocab_dict_new = {}
for k in range(N):
    v = reviews_bow[k]
    for k2,v2 in v.items():
        n += v2
        if k2 not in vocab_dict_new:
            vocab_dict_new[k2] = c
            c += 1
    new_v = {vocab_dict_new[k2]:v2 for k2,v2 in v.items()}
    reviews_bow[k] = new_v
print("Total number of words in the corpus =", n)
```

```python
all_vocabulary_dict = pickle.load(open("/content/drive/MyDrive/all_vocabulary_dict.pickle", "rb"))
all_vocabulary_dict_inv = {v:k for k,v in all_vocabulary_dict.items()}
```

```python
vocab_dict_inv = {v:all_vocabulary_dict_inv[k] for k,v in vocab_dict_new.items()}
vocab_dict = {v:k for k,v in vocab_dict_inv.items()}
pickle.dump(vocab_dict, open("/content/drive/MyDrive/vocabulary_dict.pickle", "wb"))
```

```python
vocab = set(vocab_dict.keys())
print("Number of words in the vocabulary =", len(vocab))
pickle.dump(vocab, open("/content/drive/MyDrive/vocabulary.pickle", "wb"))
```

```python
empty_documents = set()
for j in range(N):
    if len(reviews_bow[j]) == 0:
      empty_documents.add(j)
pickle.dump(empty_documents, open("/content/drive/MyDrive/empty_documents.pickle", "wb"))
print("Total number of nonempty documents =", N-len(empty_documents))
```
