```python
import pickle
```

```python
empty_documents = pickle.load(open("data/amazon_data/empty_documents.pickle", "rb"))
```

```python
reviews = pickle.load(open("data/amazon_data/reviews_filtered.pickle", "rb"))
reviews = [v for j,v in enumerate(reviews) if j not in empty_documents]
pickle.dump(reviews, open("data/amazon_data/cleaned_reviews.pickle", "wb"))
```

```python
for v in ['ratings', 'titles', 'books', 'timestamps', 'upvotes']:
    l = pickle.load(open("data/amazon_data/{}.pickle".format(v), "rb"))
    l = [e for j,e in enumerate(l) if j not in empty_documents]
    assert len(l) == len(reviews)
    pickle.dump(l, open("data/amazon_data/{}.pickle".format(v), "wb"))
```
