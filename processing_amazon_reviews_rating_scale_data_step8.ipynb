```python
import pickle
import glob
import numpy as np
```

```python
m = 10000
all_tf_files = glob.glob("/content/drive/MyDrive/fragments/tfidf_new*.pickle")
for fn in all_tf_files:
    indx = int(fn.split("/")[-1].split(".")[0][9:])
    print(indx)
    tfidf_new = pickle.load(open(fn, "rb"))
    word_counts_by_document = pickle.load(open(fn.replace("tfidf_new", "wc"), "rb"))
    word_counts_by_document_new = [None] * len(tfidf_new)
    for k,v in tfidf_new.items():
        new_k = k % m
        word_counts_by_document_new[new_k] = {k2:word_counts_by_document[new_k][k2] for k2,v2 in v.items()}
    pickle.dump(word_counts_by_document_new, open(fn.replace("tfidf_new", "wc_new"), "wb"))
```
